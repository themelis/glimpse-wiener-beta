```python
x = fits.open('../xmpl/mice_2389_catalogue_no_mask_dec2_forth2.fits')
x.info()
x[1].columns.names
```

```python
# load data from catalogue
kappa_cat = x[1].data.field('kappa')
ra=x[1].data.field('ra_gal')
dec = x[1].data.field('dec_gal')

print(ra.shape)

print(kappa_cat.shape)
```

```python
# calculate true kappa map from catalogue
npix = 256
kappa_weighted = plt.hist2d(ra, (np.rad2deg(np.sin(np.deg2rad(dec)))),
                            weights = ((kappa_cat)),
                            bins = (npix,npix), cmap = 'inferno')[0]
plt.title('weighted kappa map')

# number of background galaxies
plt.figure()
number_map = plt.hist2d(ra, (np.rad2deg(np.sin(np.deg2rad(dec)))),
                        bins = (npix, npix), cmap = 'inferno')[0]
plt.title('number of background galaxies')


kappa_map_true = np.divide(kappa_weighted, number_map)

print(kappa_map_true.shape)
```

```python
## compute the power spectrum of wt

def radial_profile(data):
    center = data.shape[0]/2
    y, x = np.indices((data.shape))
    r = np.sqrt((x - center)**2 + (y - center)**2)
    r = r.astype(np.int)

    tbin = np.bincount(r.ravel(), data.ravel())
    nr = np.bincount(r.ravel())
    radialprofile = tbin / nr
    return radialprofile

# wt_099 = np.sqrt(2) * np.ones((15,15))

fft2D = np.fft.fft2(resreal)
psd2D = np.real(fft2D*np.conj(fft2D))
psd1D = radial_profile(np.fft.fftshift(psd2D))

plt.figure(figsize = (16,8))
plt.plot(psd1D / 256)
plt.yscale('log')
plt.show()

psd1D.shape
print(np.mean(psd1D[50:100]) )
```

```python

npix = 4
P = np.zeros((npix, npix), dtype=np.complex)
freqFactor = 1.0 / npix;
for y in np.arange(npix):
    k2 = y  * freqFactor
    for x in np.arange(npix):
        k1 = x* freqFactor
        
        
        
        c1 = k1 * k1 - k2 * k2
        c2 = 2.0 * k1 * k2;
        P[y,x] = (c1 + 1j * c2)
        
print(P)



print('--')

(nx,ny) = P.shape

kl1,kl2 = np.meshgrid(np.fft.fftfreq(nx),np.fft.fftfreq(ny))

print('kl1 = \n', kl1)
print('kl2 = \n', kl2)

p1 = kl1 * kl1 - kl2 * kl2 
p2 = 2.0 * kl1 * kl2 

Pnew = p1 + 1j * p2

print(Pnew)
print('=====')
print(np.conj(Pnew) @ Pnew)

```
