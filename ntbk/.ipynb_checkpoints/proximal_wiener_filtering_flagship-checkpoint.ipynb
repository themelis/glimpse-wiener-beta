```python
# load survey

# masked
x = fits.open('../data/flagship/flagship_cat_2deg_3.fits')

x.info()
print(x[1].columns.names)

#pixels of convergence map per spatial dimension
npix = 256
    
# read survey data 
kappa_cat = x[1].data.field('kappa').astype(float)
ra=x[1].data.field('ra')
dec = x[1].data.field('dec')
e1 = -x[1].data.field('e1_noisy').astype(float)
e2 =  x[1].data.field('e2_noisy').astype(float)

# calculate true kappa map from catalogue
fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
kappa_weighted, _, _, cax1 = ax1.hist2d(ra, (np.rad2deg(np.sin(np.deg2rad(dec)))), weights = ((kappa_cat)),bins = (npix,npix), cmap = 'inferno') # , vmin= -.1, vmax = .1
ax1.set_title('kappa from survey'), plt.colorbar(cax1, ax=ax1)
number_map, _, _, cax2 = plt.hist2d(ra, (np.rad2deg(np.sin(np.deg2rad(dec)))), bins = (npix, npix), cmap = 'inferno')
ax2.set_title('number of background galaxies'), plt.colorbar(cax2, ax=ax2)

# this division insert nans because many pixels have zero galaxies
kappa_map_true = np.divide(kappa_weighted, number_map, out=np.zeros_like(kappa_weighted), where=number_map!=0)

# calculate the true g1 and g2 maps from the calalogue for KS
fig2, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
e1_bin, _, _, cax1 = ax1.hist2d(ra, (np.rad2deg(np.sin(np.deg2rad(dec)))), weights = ((e1)),bins = (npix,npix), cmap = 'inferno')
ax1.set_title('gamma1 map'), plt.colorbar(cax1, ax=ax1)
e2_bin, _, _, cax2 = ax2.hist2d(ra, (np.rad2deg(np.sin(np.deg2rad(dec)))), weights = ((e2)),bins = (npix,npix), cmap = 'inferno')
ax2.set_title('gamma2 map'), plt.colorbar(cax2, ax=ax2)

# this division inserts nans because many pixels have zero galaxies
e1_map_true = np.divide(e1_bin, number_map, out=np.zeros_like(e1_bin), where=number_map!=0) 
e2_map_true = np.divide(e2_bin, number_map, out=np.zeros_like(e2_bin), where=number_map!=0) 
```

```python
N_diagonal_from_survey = np.divide(np.var(np.concatenate((e1,e2))), number_map, out=np.zeros_like(number_map), where=number_map!=0) 
# index = np.reshape(number_map,-1)
# index = np.where(index==0)
N_diagonal_from_survey[N_diagonal_from_survey==0] = 1e19
# print(N_diagonal_from_survey==0)
# N_diagonal_from_survey[index] = 1e9
N_diagonal_from_survey.shape
```

```python
# compute noise covariance from data 

ngal = e1.shape[0]
Niter = 3 # this will take a while...
Nsvec = np.zeros((npix**2, Niter)) + 1j * np.zeros((npix**2, Niter))
for m in range(Niter):

    # permutate shear values
    rnx = np.random.permutation(ngal)
    g1p = e1[rnx]
    g2p = e2[rnx]

    # create respective noisy maps
    shear_weighted_a, _, _, cax1 = ax1.hist2d(ra, (np.rad2deg(np.sin(np.deg2rad(dec)))), weights = ((g1p)),bins = (npix,npix), cmap = 'inferno')
    shear_weighted_b, _, _, cax2 = ax2.hist2d(ra, (np.rad2deg(np.sin(np.deg2rad(dec)))), weights = ((g2p)),bins = (npix,npix), cmap = 'inferno')
    ni_real = np.divide(shear_weighted_a, number_map, out=np.zeros_like(shear_weighted_a), where=number_map!=0)
    ni_imag = np.divide(shear_weighted_b, number_map, out=np.zeros_like(shear_weighted_b), where=number_map!=0)
    
    # vectorize them
    Nsvec[:,m] = np.reshape(ni_real,-1) + 1j * np.reshape(ni_imag,-1)

# compute the covariance matrix in vectorized form
N_diagonal = np.zeros((npix * npix))
for m in range(Niter):
    for i in range(npix * npix):
        N_diagonal[i] += Nsvec[i,m].real**2 + Nsvec[i,m].imag**2 # np.outer(Nvec[:,m],np.conj(Nvec[:,m])).real # 
N_diagonal = N_diagonal / (Niter - 1)

# set high variance to empty pixels
N_diagonal[N_diagonal==0] = 1e9



```

# Apply messenger field

## Iteratively
1. $ \mathbf{t} = (\lambda \mathbf{T}) (\bar{\mathbf{N}} + \lambda \mathbf{T})^{-1} \mathbf{d} + \bar{\mathbf{N}} (\bar{\mathbf{N}} +  \lambda \mathbf{T})^{-1} \mathbf{s} $

2. $\tilde{\mathbf{s}} = \tilde{\mathbf{S}} (\tilde{\mathbf{S}} + \lambda \tilde{\mathbf{T}})^{-1} \tilde{\mathbf{t}}$

where $\bar{\mathbf{N}} = \mathbf{N} - \mathbf{T}$,  $\mathbf{T} = \tau \mathbf{1}$ and $\tau = min(diag(\mathbf{N}))$



```python
# compute wiener coefficients

tau = np.min(N_diagonal)
T_diag = tau * np.ones(npix*npix)
Tft_diag = np.ones(npix*npix) * tau**2 * float(npix*npix) # use tau**2 instead of tau

t_data_operation = T_diag / N_diagonal
t_signal_operation = (N_diagonal - T_diag) / (N_diagonal)

Sft_diagonal = np.reshape(power_map, -1)
Sft_plus_Tft_inv = Sft_diagonal / (Sft_diagonal + Tft_diag)
```

```python
number_of_iterations = 5000
s = np.zeros(npix*npix)

data_vector = np.reshape(e1_bin + 1j * e2_bin, -1)

# variance_list = []

for iteration_number in range(number_of_iterations):
    
    t = t_data_operation * data_vector +  t_signal_operation * s
    t = np.reshape(np.fft.fftshift(np.fft.fft2(np.reshape(t, (npix,npix)))), -1)
    s = Sft_plus_Tft_inv * t
    s = np.reshape(np.fft.ifft2(np.fft.fftshift(np.reshape(s, (npix,npix)))), -1)
#     variance_list.append(np.var(s))

map_wiener = np.reshape(s.real, (npix,npix))
```

```python
def radial_profile(data):
    center = data.shape[0]/2
    y, x = np.indices((data.shape))
    r = np.sqrt((x - center)**2 + (y - center)**2)
    r = r.astype(np.int)

    tbin = np.bincount(r.ravel(), data.ravel())
    nr = np.bincount(r.ravel())
    radialprofile = tbin / nr
    return radialprofile 
```

```python
#  compute power spectrum and histogram of wiener map
fft2D = np.fft.fft2(map_wiener)
psd2D = np.real(fft2D*np.conj(fft2D))
psd1D_kwm = radial_profile(np.fft.fftshift(psd2D))

fig1, ax = plt.subplots(nrows=1, ncols=2)
cax1 = ax[0].plot(psd1D_kwm), ax[0].set_title(r'$P(k)$ wiener') 
ax[0].set_yscale('log')
cax2 = ax[1].hist(np.reshape(map_wiener,-1),bins=256, histtype = 'step')
ax[1].set_title('1d hist kappa wiener')
plt.show()
```
