# Proximal Wiener filtering, toy example for validation

```python
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt

from astropy.io import fits
from scipy import ndimage
import math
```

```python
plt.rcParams['figure.figsize'] = (14, 5)
plt.rcParams.update({'font.size': 13})

# function to compute the power spectrum of a 2d image
def radial_profile(data):
    center = data.shape[0]/2
    y, x = np.indices((data.shape))
    r = np.sqrt((x - center)**2 + (y - center)**2)
    r = r.astype(np.int)

    tbin = np.bincount(r.ravel(), data.ravel())
    nr = np.bincount(r.ravel())
    radialprofile = tbin / nr
    return radialprofile / data.shape[0]**2
```

```python

# define the kappa to shear operator
def H_operator(ka_map, kb_map):
    
    # ka_map and kb_map should be of the same size
    [nx,ny] = ka_map.shape
    
    ka_map_fft = np.fft.fft2(ka_map)
    kb_map_fft = np.fft.fft2(kb_map)
    
    f1, f2 = np.meshgrid(np.fft.fftfreq(ny),np.fft.fftfreq(nx))
    
    p1 = f1 * f1 - f2 * f2
    p2 = 2 * f1 * f2
    f2 = f1 * f1 + f2 * f2
    f2[0,0] = 1 # avoid division with zero
    kafc =  (p1 * ka_map_fft - p2 * kb_map_fft) / f2
    kbfc =  (p1 * kb_map_fft + p2 * ka_map_fft) / f2
    
    g1_map = np.fft.ifft2(kafc).real
    g2_map = np.fft.ifft2(kbfc).real
    
    return g1_map, g2_map


# define the shear to convergence operator
def H_adjoint(g1_map, g2_map):
    
    [nx,ny] = g1_map.shape
    
    g1_map_ifft = np.fft.ifft2(g1_map)
    g2_map_ifft = np.fft.ifft2(g2_map)
    
    f1, f2 = np.meshgrid(np.fft.fftfreq(ny),np.fft.fftfreq(nx))
    
    p1 = f1 * f1 - f2 * f2
    p2 = 2 * f1 * f2
    f2 = f1 * f1 + f2 * f2
    f2[0,0] = 1
    g1fc =  (p1 * g1_map_ifft + p2 * g2_map_ifft) / f2
    g2fc =  (p1 * g2_map_ifft - p2 * g1_map_ifft) / f2
    
    kappa1 = np.fft.fft2(g1fc).real
    kappa2 = np.fft.fft2(g2fc).real
    
    return kappa1, kappa2

```

```python

# function to compute an 1d gaussian power spectrum
def power_function(k, sigmasignal = 7.5, amplitude = 1e6 ):
    function =  np.exp(-float(k)*float(k)/(2.0*sigmasignal*sigmasignal))/(np.sqrt(2.0*np.pi)*(sigmasignal))
    if function <= 0.0:
        return 1e-19
    else:
        return amplitude*function
    
# function to generate a convergence field out of a gaussian spectrum
def generate_gaussian_field(size):
    
    power_map = np.zeros((size, size), dtype = float)
    k_map =  np.zeros((size, size), dtype = float)

    for (i,j), val in np.ndenumerate(power_map):

        k1 = i - size/2.0
        k2 = j - size/2.0
        k_map[i, j] = (np.sqrt(k1*k1 + k2*k2))

        if k_map[i,j]==0:
            power_map[i, j] = 1e-15
        else:
            power_map[i, j] = power_function(k_map[i, j], 10.0)
            
            
    power_sample = np.random.normal(loc=0.0, scale=np.sqrt(0.5 * np.reshape(power_map, -1))) +\
               1j *  np.random.normal(loc=0.0, scale=np.sqrt(0.5 *  np.reshape(power_map, -1)))
    power_sample = np.reshape(power_sample, (size, size))
    map_fourier = np.zeros((size,size), dtype = complex)
    
    for (i, j,), value in np.ndenumerate(map_fourier):
        n = i - int(size/2)
        m = j - int(size/2)

        if n == 0 and m ==0:
            map_fourier[i,j] = 1e-15

        elif i ==0 or j == 0:
            map_fourier[i,j] = np.sqrt(2.0)*power_sample[i,j].real

        else:
            map_fourier[i,j] = power_sample[i,j] #
            map_fourier[size - i,size - j] = np.conj(power_sample[i,j])

    map_pixel = np.fft.ifft2(np.fft.fftshift(map_fourier))
    
    return map_pixel, power_map


# npix = 128
# kmap = generate_gaussian_field(npix)

# fig1,ax = plt.subplots(nrows=1, ncols=2)
# cax0 = ax[0].imshow(kmap.real, cmap = 'inferno', interpolation = 'none')
# ax[0].set_title('kmap real'), plt.colorbar(cax0, ax=ax[0])
# cax1 = ax[1].imshow(kmap.imag, cmap = 'inferno', interpolation = 'none')
# ax[1].set_title('kmap imag'), plt.colorbar(cax1, ax=ax[1])

# plt.show()

# print(np.min(kmap.real),np.max(kmap.real))
# print(np.min(kmap.imag),np.max(kmap.imag))

```

```python
def compute_spectrum_map(Px,size):
    power_map = np.zeros((size, size), dtype = float)
    k_map =  np.zeros((size, size), dtype = float)

    for (i,j), val in np.ndenumerate(power_map):

        k1 = i - size/2.0
        k2 = j - size/2.0
        k_map[i, j] = (np.sqrt(k1*k1 + k2*k2))

        if k_map[i,j]==0:
            #print(i,j)
            power_map[i, j] = 1e-15
        else:
            #print(k_map[i, j])
            power_map[i, j] = Px[int(k_map[i, j])]
    return power_map
```

```python

def prox_wiener_filtering(gamma1, gamma2, Px_map, Ncov):
        
    # initiallize
    nx = gamma1.shape[0]
    xg = np.zeros((nx,nx)) + 1j * np.zeros((nx,nx))
    
    # find the minimum noise variance
    tau = np.min(Ncov)
    
    # set the step size
    eta = 1.83 * tau
    
    # compute signal coefficient
    Esn = eta / Ncov
    
    # calculate the wiener filter coefficients
    Wfc = Px_map / (Px_map + eta)
    
    n_iter = 500
    data = gamma1 + 1j * gamma2
    for n in range(n_iter):
        t1,t2 = H_operator(xg.real, xg.imag)       # H * xg
        tj = Esn * (data - (t1 + 1j*t2))            # eta / Sn * (y- H * xg)
        t3,t4 = H_adjoint(tj.real, tj.imag) # H^T(eta / Sn * (y- H * xg))
        t = xg + (t3 + 1j*t4)                      # xg + H^T(eta / Sn * (y- H * xg))
        
        tf = np.fft.fftshift(np.fft.fft2(t))
        xgf = Wfc * tf                             # wiener filtering in fourier space
        xg = np.fft.ifft2(np.fft.fftshift(xgf))
        
    return xg.real, xg.imag
```

```python
Niter = 1
npix = 256

kmaps = np.zeros((npix,npix,Niter)) + 1j * np.zeros((npix,npix,Niter)) 
khats = np.zeros((npix,npix,Niter)) + 1j * np.zeros((npix,npix,Niter)) 
Ncovc = .3 * np.ones((npix,npix))

# Nstd = np.random.normal(np.sqrt(npow),.001,(npix,npix))
# Ncovc[:,:,i] = Nstd**2 # store covariance matrix

for i in range(Niter):
    
    # generate a gaussian field
    kmaps[:,:,i], ps_map = generate_gaussian_field(npix)
    print(kmaps[:,:,i].real.shape)
    
    # compute its power spectrum 
    fft2D = np.fft.fft2(kmaps[:,:,i].real)
    psd1D_kmt = radial_profile(np.fft.fftshift(np.real(fft2D*np.conj(fft2D))))
    
    # ps_map = compute_spectrum_map(psd1D_kmt,npix)
    
    # convert to shear
    (g1,g2) = H_operator(kmaps[:,:,i].real, kmaps[:,:,i].imag)
        
    # generate shear noise
    e1 = np.random.normal(0,np.sqrt(Ncovc / 2.),(npix,npix))
    e2 = np.random.normal(0,np.sqrt(Ncovc / 2.),(npix,npix))
    g1 += e1
    g2 += e2
    
    kh1, kh2 = prox_wiener_filtering(g1, g2, ps_map, Ncovc)

    khats[:,:,i] = kh1 + 1j * kh2
    
fig1,ax = plt.subplots(nrows=1, ncols=2)
cax0 = ax[0].imshow(ps_map, cmap = 'inferno', interpolation = 'none')
ax[0].set_title('psmap')
plt.colorbar(cax0, ax=ax[0])
plt.show()

fig2,ax = plt.subplots(nrows=1, ncols=2)
cax0 = ax[0].imshow(g1, cmap = 'inferno', interpolation = 'none')
ax[0].set_title('shear a')
plt.colorbar(cax0, ax=ax[0])
cax1 = ax[1].imshow(g2, cmap = 'inferno', interpolation = 'none')
ax[1].set_title('shear b')
plt.colorbar(cax1, ax=ax[1])
plt.show()

fig1,ax = plt.subplots(nrows=1, ncols=2)
cax0 = ax[0].imshow(kmaps[:,:,i].real, cmap = 'inferno', interpolation = 'none')
ax[0].set_title('true kmap real')
plt.colorbar(cax0, ax=ax[0])
cax1 = ax[1].imshow(kmaps[:,:,i].imag, cmap = 'inferno', interpolation = 'none')
ax[1].set_title('true kmap imag')
plt.colorbar(cax1, ax=ax[1])
plt.show()

fig3,ax = plt.subplots(nrows=1, ncols=2)
cax0 = ax[0].imshow(kh1, cmap = 'inferno', interpolation = 'none')
ax[0].set_title('wiener kmap real')
plt.colorbar(cax0, ax=ax[0])
cax1 = ax[1].imshow(kh2, cmap = 'inferno', interpolation = 'none')
ax[1].set_title('wiener kmap imag')
plt.colorbar(cax1, ax=ax[1])
plt.show()
```

```python
# compute the sample covariance of the wienered maps

Cwf = np.zeros((npix**2, npix**2))
for i in range(Niter):
    khv = khats[:,:,i].real.flatten()
    Cwf += np.outer(khv,khv)
Cwf = Cwf / (Niter - 1)

# compute the sample signal covariance
Ck = np.zeros((npix**2, npix**2))
for i in range(Niter):
    khv = kmaps[:,:,i].real.flatten()
    Ck += np.outer(khv,khv)
Ck = Ck / (Niter - 1)

# compute the sample noise covariance
Cn = np.zeros((npix**2, npix**2))
for i in range(Niter):
    khv = Ncovc[:,:,i].flatten()
    Cn += np.outer(khv,khv)
Cn = Cn / (Niter - 1)

Cwf_true  = np.matmul(np.matmul(Ck, np.linalg.inv(Ck + Cn)), np.conj(Ck))
```

# Apply messenger field

## Iteratively
1. $ \mathbf{t} = (\lambda \mathbf{T}) (\bar{\mathbf{N}} + \lambda \mathbf{T})^{-1} \mathbf{d} + \bar{\mathbf{N}} (\bar{\mathbf{N}} +  \lambda \mathbf{T})^{-1} \mathbf{s} $

2. $\tilde{\mathbf{s}} = \tilde{\mathbf{S}} (\tilde{\mathbf{S}} + \lambda \tilde{\mathbf{T}})^{-1} \tilde{\mathbf{t}}$

where $\bar{\mathbf{N}} = \mathbf{N} - \mathbf{T}$,  $\mathbf{T} = \tau \mathbf{1}$ and $\tau = min(diag(\mathbf{N}))$



```python
# compute wiener coefficients

tau = np.min(N_diagonal)
T_diag = tau * np.ones(npix*npix)
Tft_diag = np.ones(npix*npix) * tau**2 * float(npix*npix) # use tau**2 instead of tau

t_data_operation = T_diag / N_diagonal
t_signal_operation = (N_diagonal - T_diag) / (N_diagonal)

Sft_diagonal = np.reshape(power_map, -1)
Sft_plus_Tft_inv = Sft_diagonal / (Sft_diagonal + Tft_diag)
```

```python
number_of_iterations = 5000
s = np.zeros(npix*npix)

data_vector = np.reshape(e1_bin + 1j * e2_bin, -1)

# variance_list = []

for iteration_number in range(number_of_iterations):
    
    t = t_data_operation * data_vector +  t_signal_operation * s
    t = np.reshape(np.fft.fftshift(np.fft.fft2(np.reshape(t, (npix,npix)))), -1)
    s = Sft_plus_Tft_inv * t
    s = np.reshape(np.fft.ifft2(np.fft.fftshift(np.reshape(s, (npix,npix)))), -1)
#     variance_list.append(np.var(s))

map_wiener = np.reshape(s.real, (npix,npix))
```
