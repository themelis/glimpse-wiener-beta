```python
%matplotlib inline 

import numpy as np
#import scipy.stats as stats
import matplotlib.pyplot as plt
# import healpy as hp
from astropy.io import fits
# from astropy.table import Table
# from astropy.io.fits import Column

from scipy import ndimage

plt.rcParams['figure.figsize'] = (7, 7)
plt.rcParams.update({'font.size': 18})
```

```python
x = fits.open('../mice_2389_catalogue_no_mask_dec2_forth2.fits')
x.info()
x[1].columns.names
```

```python
# load data from catalogue
kappa_cat = x[1].data.field('kappa')
ra=x[1].data.field('ra_gal')
dec = x[1].data.field('dec_gal')

print(ra.shape)

print(kappa_cat.shape)
```

```python
# calculate true kappa map from catalogue
npix = 256
kappa_weighted = plt.hist2d(ra, (np.rad2deg(np.sin(np.deg2rad(dec)))),
                            weights = ((kappa_cat)),
                            bins = (npix,npix), cmap = 'inferno')[0]
plt.title('weighted kappa map')

# number of background galaxies
plt.figure()
number_map = plt.hist2d(ra, (np.rad2deg(np.sin(np.deg2rad(dec)))),
                        bins = (npix, npix), cmap = 'inferno')[0]
plt.title('number of background galaxies')


kappa_map_true = np.divide(kappa_weighted, number_map)

print(kappa_map_true.shape)
```

```python
# load intermediate results

kglpluswf = fits.getdata('./mice_2389_kappa_glimpsepluswiener_noiseless_no_mask_dec2_forth2.fits')
kglpluswf_n = fits.getdata('./mice_2389_kappa_glimpsepluswiener_noisy_no_mask_dec2_forth2.fits')

kglimpse = fits.getdata('./mice_2389_kappa_glimpse_noiseless_no_mask_dec2_forth2.fits')
kglimpse_n = fits.getdata('./mice_2389_kappa_glimpse_noisy_no_mask_dec2_forth2.fits')

kwiener = fits.getdata('./mice_2389_kappa_wiener_noiseless_no_mask_dec2_forth2.fits')
kwiener_n = fits.getdata('./mice_2389_kappa_wiener_noisy_no_mask_dec2_forth2.fits')

fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
fig1.set_size_inches(14, 5)
cax1 = ax1.imshow(kglimpse, interpolation='none', cmap='inferno' ) # , clim = (-0.03, 0.05)
ax1.set_title('kappa glimpse noiseless')
fig1.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(kglimpse_n, interpolation='none', cmap='inferno' ) # , clim = (-0.03, 0.05)
ax2.set_title('kappa glimpse noisy')
fig1.colorbar(cax2, ax=ax2)
plt.show()

fig2, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
fig2.set_size_inches(14, 5)
cax1 = ax1.imshow(kwiener, interpolation='none', cmap='inferno' ) # , clim = (-0.03, 0.05)
ax1.set_title('kappa wiener noiseless')
fig2.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(kwiener_n, interpolation='none', cmap='inferno' ) # , clim = (-0.03, 0.05)
ax2.set_title('kappa wiener noisy')
fig2.colorbar(cax2, ax=ax2)
plt.show()

fig3, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
fig3.set_size_inches(14, 5)
cax1 = ax1.imshow(kglpluswf, interpolation='none', cmap='inferno' ) # , clim = (-0.03, 0.05)
ax1.set_title('kappa glimpse plus wiener noiseless')
fig3.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(kglpluswf_n, interpolation='none', cmap='inferno' ) # , clim = (-0.03, 0.05)
ax2.set_title('kappa glimpse plus wiener noisy')
fig3.colorbar(cax2, ax=ax2)
plt.show()
```

```python
smooth_kappa = ndimage.filters.gaussian_filter(np.nan_to_num(kappa_map_true), sigma =5)
smooth_kgls = ndimage.filters.gaussian_filter(kglimpse, sigma =5)
smooth_kwf = ndimage.filters.gaussian_filter(kwiener, sigma =5)
smooth_kglpwf = ndimage.filters.gaussian_filter(kglpluswf, sigma =5)
```

```python
fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
fig1.set_size_inches(14, 5)

cax1 = ax1.imshow(smooth_kgls, interpolation='none', cmap='inferno' , clim = (-0.03, 0.05))
fig1.colorbar(cax1, ax=ax1)
ax1.set_title('smooth kappa glimpse ')

cax2 = ax2.imshow(smooth_kappa.T, interpolation='none', cmap='inferno' , clim = (-0.03, 0.05))
fig1.colorbar(cax2, ax=ax2)
ax2.set_title('smooth kappa true')

fig2, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
fig2.set_size_inches(14, 5)

cax1 = ax1.imshow(smooth_kwf, interpolation='none', cmap='inferno' , clim = (-0.03, 0.05))
fig2.colorbar(cax1, ax=ax1)
ax1.set_title('smooth kappa wiener')

cax2 = ax2.imshow(smooth_kglpwf, interpolation='none', cmap='inferno' , clim = (-0.03, 0.05))
fig2.colorbar(cax2, ax=ax2)
ax2.set_title('smooth kappa gl plus wf')

plt.show()
```
