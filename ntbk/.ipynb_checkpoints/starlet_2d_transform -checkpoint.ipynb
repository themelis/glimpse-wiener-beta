```python

%matplotlib inline

import numpy as np
# from astropy.io import fits
from scipy import ndimage
# import math
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
plt.rcParams['figure.figsize'] = (16, 4); plt.rcParams.update({'font.size': 13});
```

```python

# B3spline function 
x = np.linspace(-2,2,100)
b3s = 1/12 * (np.abs(x-2)**3 - 4 * np.abs(x-1)**3 + 6 * np.abs(x)**3 - 4 * np.abs(x+1)**3 + np.abs(x+2)**3)
plt.plot(b3s)
plt.show()
```

```python
# # # 1st generation starlet transform 

def b3spline_conv(im_in, step_trou):

    # create the filter coefficients
    h0 = 3 / 8
    h1 = 1 / 4
    h2 = 1 / 16
    
    hlen = 4 * step_trou + 1
    h = np.zeros(hlen)
    mi = int(hlen / 2)
    h[mi] = h0
    h[mi + step_trou] = h1
    h[mi - step_trou] = h1
    h[0]  = h2
    h[-1] = h2
    
    v = np.zeros_like(im_in)
    o = np.zeros_like(im_in)
    for n in range(im_in.shape[0]): 
        # print(np.shape(im_in[n]))
        v[n,:] = np.convolve(im_in[n,:], h, mode='same') # convolve rows
    
    for n in range(im_in.shape[1]):
        o[:,n] = np.convolve(v[:,n], h, mode='same') # convolve columns
    
    return o

def star2d_transform(image, nscales, gen2):
    
    Nx, Ny = image.shape
    Nz = nscales
    wt = np.zeros((Nz+1, Nx, Ny))
    
    imin = image.copy()
    step = 1    
    for i in range(Nz):
        imc = b3spline_conv(imin, step)
        
        if gen2:
            wt[i,:,:] = imin - b3spline_conv(imc, step)
        else:
            wt[i,:,:] = imin - imc
        
        imin = imc.copy()
        
        step *= 2
    
    wt[i+1,:,:] = imc
    
    return wt


def star2d_adjoint_transform(wcoeffs, gen2):
    
    if not gen2:
        o = np.sum(wcoeffs, axis=0)
    else:
        Nz, Nx, Ny = wcoeffs.shape
        
        step_trou = np.power(2, Nz-2)
        
        o = wcoeffs[Nz-1,:,:]
        for i in reversed(range(Nz-2)):
            o = wcoeffs[i,:,:] + b3spline_conv(o, step)
            
            step_trou /= 2
    
    return o

def main():
    
    img = mpimg.imread('../data/einstein.jpg') 
    
    imdelta = np.zeros((256,256))
    imdelta[128,128] = 10
    
    plt.imshow(imdelta)
    plt.colorbar()
    plt.show()
    
    wc2d = star2d_transform(img, 5, False)
    print('wcoeffs scales: ', wc2d.shape[0])
    
    for i in range(0, wc2d.shape[0], 2):
        fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
        cax1 = ax1.imshow(wc2d[i,:,:], interpolation='none') # , cmap='gray', origin='lower' 
        ax1.set_title('scale ' + str(i)), plt.colorbar(cax1, ax=ax1)
        cax2 = ax2.imshow(wc2d[i,:,:], interpolation='none') 
        ax2.set_title('scale ' + str(i+1)), plt.colorbar(cax2, ax=ax2)

    plt.show()
    
    print('--- reconstruction test ---')
    
    imrec = np.sum(wc2d, axis=0)
    plt.imshow(imrec)
    plt.colorbar()
    plt.show()
    
    
if __name__ == '__main__':
    main()
```
