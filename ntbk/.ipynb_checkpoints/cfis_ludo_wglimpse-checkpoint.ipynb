```python
%matplotlib inline

import numpy as np
from astropy.io import fits
from scipy import ndimage
import math
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (16, 6); plt.rcParams.update({'font.size': 13});

def radial_profile(data):
    center = data.shape[0]/2
    y, x = np.indices((data.shape))
    r = np.sqrt((x - center)**2 + (y - center)**2)
    r = r.astype(np.int)

    tbin = np.bincount(r.ravel(), data.ravel())
    nr = np.bincount(r.ravel())
    radialprofile = tbin / nr
    return radialprofile / data.shape[0]**2
```

```python
# calculate the binning edges used in glimpse
pixel_size_arcmin = 3.0
npix = 256
lim = pixel_size_arcmin * (npix / 2) / 60

pixel_edges = np.deg2rad(np.linspace(-lim, lim, npix+1))

# kgl = fits.open('../data/flagship/kappa_cat_2dec_3_positiviny_false_glimpse_old.fits')
# ra1 = kgl[1].data
# dec1 = kgl[2].data

# # # compute the gnonomic projection coordinates
# ra1  *= np.pi / 180.0 # convert degrees to radian
# dec1 *= np.pi / 180.0 # convert degrees to radian
# center_ra1 = 2 * np.pi / 180.0 
# center_dec1 = 2 * np.pi / 180.0
# denom1 = np.cos(center_dec1) * np.cos(dec1) * np.cos(ra1  - center_ra1) + np.sin(center_dec1) * np.sin(dec1);
# xgn1 = np.cos(dec1) * np.sin(ra1  - center_ra1) / denom1;
# ygn1 = (np.cos(center_dec1) * np.sin(dec1) - np.cos(dec1) * np.sin(center_dec1) * np.cos(ra1 - center_ra1)) / denom1;

# fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
# cax1 = ax1.imshow(ra1, interpolation='none', cmap='inferno') 
# ax1.set_title('ra1'), plt.colorbar(cax1, ax=ax1)
# cax2 = ax2.imshow(dec1, interpolation='none', cmap='inferno') 
# ax2.set_title('dec1'), plt.colorbar(cax2, ax=ax2)
# plt.show()

```

```python
# load survey
x = fits.open('../data/cfis/new/cfis_ludo.fits')

x.info()
print(x[1].columns.names)

# npix = 256

# read survey data 
ra  =  x[1].data.field('ra_gal')
dec =  x[1].data.field('dec_gal')
e1  =  x[1].data.field('e1_gal').astype(float)
e2  =  x[1].data.field('e2_gal').astype(float)

ngal = e1.shape[0]
print('---\n#galaxies: ', ngal,'\n---\n')

# calculate the gnonomic projection coordinates
ra_rad     = ra * np.pi / 180.0 # convert degrees to radian
dec_rad    = dec * np.pi / 180.0 # convert degrees to radian
center_ra  = 214.58669326 * np.pi / 180.0 
center_dec = 54.49917587 * np.pi / 180.0
denom = np.cos(center_dec) * np.cos(dec_rad) * np.cos(ra_rad  - center_ra) + np.sin(center_dec) * np.sin(dec_rad);
xgn = np.cos(dec_rad) * np.sin(ra_rad  - center_ra) / denom;
ygn = (np.cos(center_dec) * np.sin(dec_rad) - np.cos(dec_rad) * np.sin(center_dec) * np.cos(ra_rad - center_ra)) / denom;

# # calculate true kappa map from catalogue
fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
number_map, _, _, cax2 = plt.hist2d(xgn, ygn, bins = (pixel_edges, pixel_edges), cmap = 'inferno')
ax2.set_title('number of background galaxies'), plt.colorbar(cax2, ax=ax2)
# transpose the hist2d result
number_map = number_map.T
# this division insert nans because many pixels have zero galaxies

# calculate the true g1 and g2 maps from the calalogue for KS
fig2, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
g1_bin, _, _, cax1 = ax1.hist2d(xgn, ygn, weights = e1, bins = (pixel_edges,pixel_edges), cmap = 'inferno')
ax1.set_title('gamma1 map'), plt.colorbar(cax1, ax=ax1)
g2_bin, _, _, cax2 = ax2.hist2d(xgn, ygn, weights = e2, bins = (pixel_edges,pixel_edges), cmap = 'inferno')
ax2.set_title('gamma2 map'), plt.colorbar(cax2, ax=ax2)
# transpose the hist2d result
g1_bin = g1_bin.T
g2_bin = g2_bin.T 

# this division inserts nans because many pixels have zero galaxies
g1_map_true = np.divide(g1_bin, number_map, out=np.zeros_like(g1_bin), where=number_map!=0)
g2_map_true = np.divide(g2_bin, number_map, out=np.zeros_like(g2_bin), where=number_map!=0) 
```

```python
# # # compute noise covariance from data 

Niter = 300 # this will take a while...
Nsvec = np.zeros((npix**2, Niter)) + 1j * np.zeros((npix**2, Niter))
for m in range(Niter):

    # permutate shear values
    rnx1 = np.random.permutation(e1.shape[0])
    rnx2 = np.random.permutation(e2.shape[0])
    g1p = e1[rnx1]
    g2p = e2[rnx2]

    # create respective noisy maps
    shear_weighted_a, _, _, cax1 = ax1.hist2d(xgn, ygn, weights = g1p, bins = (pixel_edges,pixel_edges), cmap = 'inferno')
    shear_weighted_b, _, _, cax2 = ax2.hist2d(xgn, ygn, weights = g2p, bins = (pixel_edges,pixel_edges), cmap = 'inferno')
    shear_weighted_a = shear_weighted_a.T
    shear_weighted_b = shear_weighted_b.T
    ni_real = np.divide(shear_weighted_a, number_map, out=np.zeros_like(shear_weighted_a), where=number_map!=0)
    ni_imag = np.divide(shear_weighted_b, number_map, out=np.zeros_like(shear_weighted_b), where=number_map!=0)
    
    # vectorize them
    Nsvec[:,m] = np.reshape(ni_real,-1) + 1j * np.reshape(ni_imag,-1)

# compute the covariance matrix in vectorized form
N_diagonal = np.zeros((npix * npix))
for m in range(Niter):
    N_diagonal += Nsvec[:,m].real**2 + Nsvec[:,m].imag**2 # np.outer(Nvec[:,m],np.conj(Nvec[:,m])).real # 
N_diagonal = N_diagonal / (Niter - 1)

# set high variance to masked pixels
N_diagonal[N_diagonal==0] = 1e9
```

```python
# # # validate the noise covariance matrix

Ncov = np.reshape(N_diagonal,(npix,npix))

# compute noise covariance theoretically 
Ncov_th = np.divide(2 * np.var(e1), number_map, out=np.zeros_like(Ncov), where=number_map!=0)
Ncov_th[Ncov_th==0] = 1e9

fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(Ncov, interpolation='none', cmap='inferno', vmax = .05) 
ax1.set_title('Ncov'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(Ncov_th, interpolation='none', cmap='inferno', vmax = .05) 
ax2.set_title('Ncov_th'), plt.colorbar(cax2, ax=ax2)

plt.show()
```

## Kaiser-Squires inversion

```python

# Compute once the Fourier kernels for the transform
k1, k2 = np.meshgrid(np.fft.fftfreq(npix), np.fft.fftfreq(npix))
denom = k1*k1 + k2*k2
denom[0, 0] = 1  # avoid division by 0
kernel1 = (k1**2 - k2**2)/denom
kernel2 = (2*k1*k2)/denom

def kappa_to_gamma(kappa):
    k = np.fft.fft2(kappa)
    g1 = np.fft.ifft2(kernel1 * k) 
    g2 = np.fft.ifft2(kernel2 * k)
    return g1.real - g2.imag, g2.real + g1.imag 

def gamma_to_kappa(gam1, gam2):
    g = gam1 + 1j*gam2
    return np.fft.ifft2((kernel1 - 1j*kernel2)* np.fft.fft2(g))
```

```python
# # # compute the Kaiser-Squares map

kgka = gamma_to_kappa(g1_map_true, g2_map_true)
kgkas = ndimage.filters.gaussian_filter(kgka.real, sigma = 2)
```

```python
# # # compare wglimpse map with smooth KS

kfgl = fits.getdata('../data/cfis/new/cfis_ludo_kappa_lambda3_positivityfalse.fits')

kfgl = kfgl[16:-16,16:-16]
kgkas = kgkas[16:-16,16:-16]
vvmin = np.min(kgkas)
vvmax = np.max(kgkas)
fig2, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(kgkas, interpolation='none', cmap='inferno',origin='lower') 
ax1.set_title('kappa smooth Kaiser-Squares'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(kfgl, interpolation='none', cmap='inferno',origin='lower') 
ax2.set_title('kappa with mask true'), plt.colorbar(cax2, ax=ax2)

plt.show()
```
