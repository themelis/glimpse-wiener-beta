```python
%matplotlib inline

import numpy as np
from astropy.io import fits
from scipy import ndimage
import math
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (16, 6); plt.rcParams.update({'font.size': 13});
```

```python
# # # open toy dataset 
toy = fits.open('../data/toy/cat_toy_4.fits')
print(toy.info())
print(toy[1].columns.names)

# read survey data 
ra  = toy[1].data.field('ra_gal')
dec = toy[1].data.field('dec_gal')
e1  = toy[1].data.field('e1_gal').astype(float)
e2  = toy[1].data.field('e2_gal').astype(float)
z   = toy[1].data.field('z_gal').astype(float)
ktr = toy[1].data.field('kappa').astype(float)

ngal = e1.shape[0]

print('---\n#galaxies: ', ngal,'\n---')

```

```python
# calculate the binning edges used in glimpse
npix = 128
pixel_size_arcmin = .46875
lim = pixel_size_arcmin * (npix / 2) / 60
pixel_edges = np.deg2rad(np.linspace(-lim, lim, npix+1))

# calculate the gnonomic projection coordinates
ra_rad     = ra * np.pi / 180.0 # convert degrees to radian
dec_rad    = dec * np.pi / 180.0 # convert degrees to radian
center_ra  = 0 
center_dec = 0 
denom = np.cos(center_dec) * np.cos(dec_rad) * np.cos(ra_rad  - center_ra) + np.sin(center_dec) * np.sin(dec_rad)
xgn = np.cos(dec_rad) * np.sin(ra_rad  - center_ra) / denom
ygn = (np.cos(center_dec) * np.sin(dec_rad) - np.cos(dec_rad) * np.sin(center_dec) * np.cos(ra_rad - center_ra)) / denom

# # calculate true kappa map from catalogue
fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
kappa_weighted, _, _, cax1 = ax1.hist2d(xgn, ygn, weights = ktr, bins = (pixel_edges,pixel_edges))
ax1.set_title('kappa from survey'), plt.colorbar(cax1, ax=ax1)
number_map, _, _, cax2 = plt.hist2d(xgn, ygn, bins = (pixel_edges, pixel_edges))
ax2.set_title('number of background galaxies'), plt.colorbar(cax2, ax=ax2)
plt.close()

# transpose the hist2d result
number_map = number_map.T
kappa_weighted = kappa_weighted.T
# this division insert nans because many pixels have zero galaxies
ktr_map = np.divide(kappa_weighted, number_map, out=np.zeros_like(kappa_weighted), where=number_map!=0)

# calculate the true g1 and g2 maps from the calalogue for KS
fig2, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
g1_bin, _, _, cax1 = ax1.hist2d(xgn, ygn, weights = e1, bins = (pixel_edges,pixel_edges))
ax1.set_title(r'$\gamma_1$ map'), plt.colorbar(cax1, ax=ax1)
g2_bin, _, _, cax2 = ax2.hist2d(xgn, ygn, weights = e2, bins = (pixel_edges,pixel_edges))
ax2.set_title('$\gamma_2$ map'), plt.colorbar(cax2, ax=ax2)
plt.close()
# transpose the hist2d result
g1_bin = g1_bin.T
g2_bin = g2_bin.T 

# this division inserts nans because many pixels have zero galaxies
g1_map_true = np.divide(g1_bin, number_map, out=np.zeros_like(g1_bin), where=number_map!=0)
g2_map_true = np.divide(g2_bin, number_map, out=np.zeros_like(g2_bin), where=number_map!=0) 
```

```python
fig2, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(g1_map_true, cmap = 'inferno')
ax1.set_title(r'$\gamma_1$ map'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(g2_map_true, cmap = 'inferno')
ax2.set_title(r'$\gamma_2$ map'), plt.colorbar(cax2, ax=ax2)

fig2, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(ktr_map, cmap = 'inferno')
ax1.set_title(r'true $\kappa$ map'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(number_map, cmap = 'inferno')
ax2.set_title('galaxy density map'), plt.colorbar(cax2, ax=ax2)

plt.show()
```

```python
# # # compute the noise covariance matrix

ra_l_index = np.where(ra<=0)
ra_r_index = np.where(ra>0)

eln = np.array([e1[ra_l_index], e2[ra_l_index]])
ehn = np.array([e1[ra_r_index], e2[ra_r_index]])

print(np.var(eln))
print(np.var(ehn))

Nc_l = np.var(eln) * np.ones((128,64))
Nc_r = np.var(ehn) * np.ones((128,64))

Nc_full = np.concatenate((Nc_l,Nc_r), axis=1)

Ncov_th = np.divide(2 * Nc_full, number_map, out=np.zeros_like(number_map), where=number_map!=0)
Ncov_th[Ncov_th==0] = 1e9

fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(Nc_full, interpolation='none', cmap='inferno') 
ax1.set_title('Nc_full'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(Ncov_th, interpolation='none', cmap='inferno', vmax = .1) 
ax2.set_title('Ncov_th'), plt.colorbar(cax2, ax=ax2)

plt.show()

```

## Kaiser-Squires inversion

```python

# Compute once the Fourier kernels for the transform
k1, k2 = np.meshgrid(np.fft.fftfreq(npix), np.fft.fftfreq(npix))
denom = k1*k1 + k2*k2
denom[0, 0] = 1  # avoid division by 0
kernel1 = (k1**2 - k2**2)/denom
kernel2 = (2*k1*k2)/denom

def kappa_to_gamma(kappa):
    k = np.fft.fft2(kappa)
    g1 = np.fft.ifft2(kernel1 * k) 
    g2 = np.fft.ifft2(kernel2 * k)
    return g1.real - g2.imag, g2.real + g1.imag 

def gamma_to_kappa(gam1, gam2):
    g = gam1 + 1j*gam2
    return np.fft.ifft2((kernel1 - 1j*kernel2)* np.fft.fft2(g))
```

```python
# # # compute the Kaiser-Squares map

kks = gamma_to_kappa(g1_map_true, g2_map_true)
kkss = ndimage.filters.gaussian_filter(kks.real, sigma = 2)
```

```python
# # # KS map

vvmin = np.min(kkss)
vvmax = np.max(kkss)
fig2, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(kks.real, interpolation='none', cmap='inferno', vmin = vvmin, vmax = vvmax) 
ax1.set_title('kappa Kaiser-Squares'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(kkss, interpolation='none', cmap='inferno', vmin = vvmin, vmax = vvmax) 
ax2.set_title('smooth'), plt.colorbar(cax2, ax=ax2)

plt.show()
```
