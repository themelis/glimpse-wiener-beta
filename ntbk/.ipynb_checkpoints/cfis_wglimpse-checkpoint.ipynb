# wglimpse for CFIS data

```python

%matplotlib inline

import numpy as np
from astropy.io import fits
from scipy import ndimage
import math
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (16, 6); plt.rcParams.update({'font.size': 13});
```

```python
# get the glimpse input shear values 
e1_map = fits.getdata('../data/cfis/e1map.fits')
e2_map = fits.getdata('../data/cfis/e2map.fits')

fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(e1_map, interpolation='none', cmap='inferno') 
ax1.set_title('e1 map'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(e2_map, interpolation='none', cmap='inferno') 
ax2.set_title('e2 map'), plt.colorbar(cax2, ax=ax2)
plt.show()

# get the noise covariance

Ncov = fits.getdata('../data/cfis/noise_cov.fits')
fig2, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(Ncov, interpolation='none', cmap='inferno') 
ax1.set_title('noise covariance map'), plt.colorbar(cax1, ax=ax1)
plt.show()
```

```python

# Compute once the Fourier kernels for the transform
npix = e1_map.shape[0]
k1, k2 = np.meshgrid(np.fft.fftfreq(npix), np.fft.fftfreq(npix))
denom = k1*k1 + k2*k2
denom[0, 0] = 1  # avoid division by 0
kernel1 = (k1**2 - k2**2)/denom
kernel2 = (2*k1*k2)/denom

def kappa_to_gamma(kappa):
    k = np.fft.fft2(kappa)
    g1 = np.fft.ifft2(kernel1 * k) 
    g2 = np.fft.ifft2(kernel2 * k)
    return g1.real - g2.imag, g2.real + g1.imag 

def gamma_to_kappa(gam1, gam2):
    g = gam1 + 1j*gam2
    return np.fft.ifft2((kernel1 - 1j*kernel2)* np.fft.fft2(g))


# plt.imshow(kgkas, interpolation='none', cmap='inferno', origin='lower') 
# plt.title('Kaiser Squires')
# plt.colorbar()
# plt.show()
```

```python

# compute the Kaiser-Squares map
kgka = gamma_to_kappa(e1_map, e2_map)

# smooth out the KS maps 
kgkas = ndimage.filters.gaussian_filter(kgka.real, sigma = 3)
```

```python
# get the glimpse kappa map
kgl = fits.getdata('../data/cfis/kappa_cfis.fits')

# plot results
fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(kgl, interpolation='none', cmap='inferno',origin='lower') 
ax1.set_title('new glimpse'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(kgkas, interpolation='none', cmap='inferno',origin='lower') 
ax2.set_title('Kaiser Squires'), plt.colorbar(cax2, ax=ax2)
plt.show()

# plt.imshow(kgl, interpolation='none', cmap='inferno', origin='lower') 
# plt.title('new glimpse')
# plt.colorbar()
# plt.show()
```
