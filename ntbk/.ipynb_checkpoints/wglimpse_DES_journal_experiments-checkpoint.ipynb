```python
# calculate the binning edges used in glimpse
npix = 256 #  612 
pixel_size_arcmin = .079 * 60 #  .05 * 60  
lim = pixel_size_arcmin * (npix / 2) / 60
pixel_edges = np.deg2rad(np.linspace(-lim, lim, npix+1))

# [+75.0,-52.5]

# calculate the gnonomic projection coordinates
ra_rad     = ra * np.pi / 180.0 # convert degrees to radian
dec_rad    = dec * np.pi / 180.0 # convert degrees to radian
center_ra  = 76.95 * np.pi / 180.0
center_dec = -52.23 * np.pi / 180.0
denom = np.cos(center_dec) * np.cos(dec_rad) * np.cos(ra_rad  - center_ra) + np.sin(center_dec) * np.sin(dec_rad)
xgn = np.cos(dec_rad) * np.sin(ra_rad  - center_ra) / denom
ygn = (np.cos(center_dec) * np.sin(dec_rad) - np.cos(dec_rad) * np.sin(center_dec) * np.cos(ra_rad - center_ra)) / denom

# # there is no true kappa map in survey
fig2, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
number_map, _, _, cax2 = plt.hist2d(xgn, ygn, bins = (pixel_edges, pixel_edges))
ax2.set_title('number of background galaxies'), plt.colorbar(cax2, ax=ax2)
plt.close()
# transpose the hist2d result
number_map = number_map.T

# calculate the true g1 and g2 maps from the calalogue for KS
fig2, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
g1_bin, _, _, cax1 = ax1.hist2d(xgn, ygn, weights = e1, bins = (pixel_edges,pixel_edges))
ax1.set_title(r'$\gamma_1$ map'), plt.colorbar(cax1, ax=ax1)
g2_bin, _, _, cax2 = ax2.hist2d(xgn, ygn, weights = e2, bins = (pixel_edges,pixel_edges))
ax2.set_title('$\gamma_2$ map'), plt.colorbar(cax2, ax=ax2)
plt.close()
# transpose the hist2d result
g1_bin = g1_bin.T
g2_bin = g2_bin.T 

# this division inserts nans because many pixels have zero galaxies
e1_map = np.divide(g1_bin, number_map, out=np.zeros_like(g1_bin), where=number_map!=0)
e2_map = np.divide(g2_bin, number_map, out=np.zeros_like(g2_bin), where=number_map!=0) 
```

```python
# # # plot the binned shear 
fig2, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(e1_map, cmap = 'inferno')
ax1.set_title(r'$\epsilon_1$ map'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(e2_map, cmap = 'inferno')
ax2.set_title(r'$\epsilon_2$ map'), plt.colorbar(cax2, ax=ax2)

plt.show()
```

```python
# # # compute power spectrum and noise covariance 

ps1d = fits.getdata('../data/DES/fits_files/DES_ps1d.fits')

# compute noise covariance from data 
Niter = 300 # this will take a while...
Nsvec = np.zeros((npix**2, Niter)) + 1j * np.zeros((npix**2, Niter))
for m in range(Niter):

    # permutate shear values
    rnx1 = np.random.permutation(e1.shape[0])
    rnx2 = np.random.permutation(e2.shape[0])
    g1p = e1[rnx1]
    g2p = e2[rnx2]

    # create respective noisy maps
    shear_weighted_a, _, _, cax1 = ax1.hist2d(xgn, ygn, weights = g1p, bins = (pixel_edges,pixel_edges))
    shear_weighted_b, _, _, cax2 = ax2.hist2d(xgn, ygn, weights = g2p, bins = (pixel_edges,pixel_edges))
    shear_weighted_a = shear_weighted_a.T
    shear_weighted_b = shear_weighted_b.T
    ni_real = np.divide(shear_weighted_a, number_map, out=np.zeros_like(shear_weighted_a), where=number_map!=0)
    ni_imag = np.divide(shear_weighted_b, number_map, out=np.zeros_like(shear_weighted_b), where=number_map!=0)
    
    # vectorize them
    Nsvec[:,m] = np.reshape(ni_real,-1) + 1j * np.reshape(ni_imag,-1)

# compute the covariance matrix in vectorized form
N_diagonal = np.zeros((npix * npix))
for m in range(Niter):
    N_diagonal += Nsvec[:,m].real**2 + Nsvec[:,m].imag**2 # np.outer(Nvec[:,m],np.conj(Nvec[:,m])).real # 
N_diagonal = N_diagonal / (Niter - 1)

# set high variance to masked pixels
N_diagonal[N_diagonal==0] = 1e9

Ncov = np.reshape(N_diagonal,(npix,npix))

unmasked_pixels_index = np.where(Ncov<1e2)
```

```python
def radial_profile(data):
    center = data.shape[0]/2
    y, x = np.indices((data.shape))
    r = np.sqrt((x - center)**2 + (y - center)**2)
    r = r.astype(np.int)

    tbin = np.bincount(r.ravel(), data.ravel())
    nr = np.bincount(r.ravel())
    radialprofile = tbin / nr
    return radialprofile / data.shape[0]**2
```

### Kaiser-Squires inversion

```python

# Compute once the Fourier kernels for the transform
k1, k2 = np.meshgrid(np.fft.fftfreq(npix), np.fft.fftfreq(npix))
denom = k1*k1 + k2*k2
denom[0, 0] = 1  # avoid division by 0
kernel1 = (k1**2 - k2**2)/denom
kernel2 = (2*k1*k2)/denom

def kappa_to_gamma(kappa):
    k = np.fft.fft2(kappa)
    g1 = np.fft.ifft2(kernel1 * k) 
    g2 = np.fft.ifft2(kernel2 * k)
    return g1.real - g2.imag, g2.real + g1.imag 

def gamma_to_kappa(gam1, gam2):
    g = gam1 + 1j*gam2
    return np.fft.ifft2((kernel1 - 1j*kernel2)* np.fft.fft2(g))
```

```python
# # # compute the Kaiser-Squares map

kks = gamma_to_kappa(e1_map, e2_map)
kkss = ndimage.filters.gaussian_filter(kks.real, sigma = 2)

kkss_mask = np.zeros_like(kkss)
kkss_mask[unmasked_pixels_index] = kkss[unmasked_pixels_index]
kkss_mask = np.flip(kkss_mask, 1)
```

### Wiener filtering

```python
# wiener filtering inversion

# define the kappa to shear operator
def H_operator(ka_map, kb_map):
    
    # ka_map and kb_map should be of the same size
    [nx,ny] = ka_map.shape
    
    ka_map_fft = np.fft.fft2(ka_map)
    kb_map_fft = np.fft.fft2(kb_map)
    
    f1, f2 = np.meshgrid(np.fft.fftfreq(ny),np.fft.fftfreq(nx))
    
    p1 = f1 * f1 - f2 * f2
    p2 = 2 * f1 * f2
    f2 = f1 * f1 + f2 * f2
    f2[0,0] = 1 # avoid division with zero
    kafc =  (p1 * ka_map_fft - p2 * kb_map_fft) / f2
    kbfc =  (p1 * kb_map_fft + p2 * ka_map_fft) / f2
    
    g1_map = np.fft.ifft2(kafc).real
    g2_map = np.fft.ifft2(kbfc).real
    
    return g1_map, g2_map


# define the shear to convergence operator
def H_adjoint(g1_map, g2_map):
    
    [nx,ny] = g1_map.shape
    
    g1_map_ifft = np.fft.ifft2(g1_map)
    g2_map_ifft = np.fft.ifft2(g2_map)
    
    f1, f2 = np.meshgrid(np.fft.fftfreq(ny),np.fft.fftfreq(nx))
    
    p1 = f1 * f1 - f2 * f2
    p2 = 2 * f1 * f2
    f2 = f1 * f1 + f2 * f2
    f2[0,0] = 1
    g1fc =  (p1 * g1_map_ifft + p2 * g2_map_ifft) / f2
    g2fc =  (p1 * g2_map_ifft - p2 * g1_map_ifft) / f2
    
    kappa1 = np.fft.fft2(g1fc).real
    kappa2 = np.fft.fft2(g2fc).real
    
    return kappa1, kappa2

def compute_spectrum_map(Px,size):
    power_map = np.zeros((size, size), dtype = float)
    k_map =  np.zeros((size, size), dtype = float)

    for (i,j), val in np.ndenumerate(power_map):

        k1 = i - size/2.0
        k2 = j - size/2.0
        k_map[i, j] = (np.sqrt(k1*k1 + k2*k2))

        if k_map[i,j]==0:
            #print(i,j)
            power_map[i, j] = 1e-15
        else:
            #print(k_map[i, j])
            power_map[i, j] = Px[int(k_map[i, j])]
    return power_map


def prox_wiener_filtering(gamma1, gamma2, Px_map, Ncv):
        
    # initiallize
    nx = gamma1.shape[0]
    xg = np.zeros((nx,nx)) + 1j * np.zeros((nx,nx))
    
    # find the minimum noise variance
    tau = np.min(Ncv)
    
    # set the step size
    eta = 1.83 * tau
    
    # compute signal coefficient
    Esn = eta / Ncv
    
    # calculate the wiener filter coefficients
    Wfc = Px_map / (Px_map + eta)
    
    n_iter = 100
    for n in range(n_iter):
        t1,t2 = H_operator(xg.real, xg.imag)                              # H * xg
        t1,t2 = H_adjoint(Esn*(gamma1 - t1), Esn*(gamma2 - t2))           # H^T(eta / Sn * (y- H * xg))
        t = xg + (t1 + 1j*t2)                                             # xg + H^T(eta / Sn * (y- H * xg))
        
        xg = np.fft.ifft2(np.fft.fftshift(Wfc * np.fft.fftshift(np.fft.fft2(t)))) # wiener filtering in fourier space
        
    return xg.real, xg.imag

(kpwa,kpwb) = prox_wiener_filtering(e1_map, e2_map, compute_spectrum_map(ps1d_DES,npix), Ncov)

kwr_mask = np.zeros_like(kpwa)
kwr_mask[unmasked_pixels_index] = kpwa[unmasked_pixels_index]
kwr_mask = np.flip(kwr_mask, 1)
```
