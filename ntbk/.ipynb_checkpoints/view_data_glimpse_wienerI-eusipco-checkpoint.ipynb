```python
kwiener = fits.getdata('../build/kwiener.fits') 
kgs = fits.getdata('../example/masked_mice_2389_catalogue_small_corrected_kappa_glimpse.fits') 
kappa_from_data = fits.getdata('../example/2713_kappa_true.fits') 
kglimpse = fits.getdata('../build/kglimpse.fits') 
glpluswf = fits.getdata('../build/glpluswf.fits')

fig5, [ax1, ax2] = plt.subplots(nrows=1, ncols=2) 
fig5.set_size_inches(14, 5)

cax1 = ax1.imshow(kgs, interpolation='none', cmap='inferno') 
fig5.colorbar(cax1, ax = ax1) 
ax1.set_title('kappa glimpse')

cax2 = ax2.imshow(kappa_from_data.T, interpolation='none', cmap='inferno' ) 
fig5.colorbar(cax2, ax = ax2) 
ax2.set_title('kappa true') 
plt.show()

fig4, [ax1, ax2] = plt.subplots(nrows=1, ncols=2) 
fig4.set_size_inches(14, 5)

cax1 = ax1.imshow(kwiener, interpolation='none', cmap='inferno' ) 
fig4.colorbar(cax1, ax = ax1) 
ax1.set_title('kappa wiener')

cax2 = ax2.imshow(kgs - kwiener, interpolation='none', cmap='inferno') 
fig4.colorbar(cax2, ax = ax2) 
ax2.set_title('difference')

plt.show()
```

```python
## compute the power spectrum of wt

def radial_profile(data):
    center = data.shape[0]/2
    y, x = np.indices((data.shape))
    r = np.sqrt((x - center)**2 + (y - center)**2)
    r = r.astype(np.int)

    tbin = np.bincount(r.ravel(), data.ravel())
    nr = np.bincount(r.ravel())
    radialprofile = tbin / nr
    return radialprofile

fft2D = np.fft.fft2(wt_099)
psd2D = np.real(fft2D*np.conj(fft2D))
psd1D = radial_profile(np.fft.fftshift(psd2D))

plt.figure(figsize = (16,8))
plt.plot(psd1D)
plt.yscale('log')
plt.show()

psd1D.shape
```
