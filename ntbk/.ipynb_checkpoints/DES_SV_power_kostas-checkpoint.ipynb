# This notebook reads cosmosis parameters and generates convergence maps that simulate mice data. Different realizations are used to compute the power spectrum of a mice convergence map

```python
%matplotlib inline 

import numpy as np
import time
import math
import healpy as hp
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import os
import scipy.ndimage as ndimage
from scipy.interpolate import interp1d
from astropy.io import fits
from astropy.table import Table
from astropy.io.fits import Column

```

```python
plt.rcParams['figure.figsize'] = (14, 4)
```

```python
ell_input = np.loadtxt('../data/mice/old_cosmosis_sv_nz/ell.txt')
cl_input = np.loadtxt('../data/mice/old_cosmosis_sv_nz/bin_1_1.txt')
```

```python
print(cl_input)
```

```python
_ = plt.subplot(1,2,1), plt.loglog(ell_input, cl_input)
_ = plt.grid(True), plt.ylabel(r'$C_l$', fontsize = 18), plt.xlabel(r'$l$', fontsize = 18)
_ = plt.subplot(1,2,2), plt.loglog(ell_input, cl_input*ell_input*(ell_input+1.0)/(2.0*np.pi))
_ = plt.grid(True), plt.ylabel(r'$C_l l (l + 1)/ 2 \pi$', fontsize = 18), plt.xlabel(r'$l$', fontsize = 18)
```

```python
interpolate_cl = interp1d(ell_input, cl_input)
ell_linear = range(1,10000)
```

```python
nside = 1024
kappa_map_gaussian = hp.synfast(interpolate_cl(ell_linear), nside, lmax=3*nside-1, mmax=None, pol=False, pixwin=True)
```

```python
gaussian_projected = hp.gnomview(kappa_map_gaussian, rot=[+75.0,-52.5],reso=4.5, xsize=256, ysize=256,
                                 flip ='geo', return_projected_map =True)
```

```python
gaussian_projected.shape
```

```python
def radial_profile(data):
    center = data.shape[0]/2
    y, x = np.indices((data.shape))
    r = np.sqrt((x - center)**2 + (y - center)**2)
    r = r.astype(np.int)

    tbin = np.bincount(r.ravel(), data.ravel())
    nr = np.bincount(r.ravel())
    radialprofile = tbin / nr
    return radialprofile / data.shape[0]**2
```

```python
ps1d = radial_profile(gaussian_projected)
ps1d.shape
```

```python
Niter = 2

ps1ds = np.zeros((ps1d.shape[0],Niter*17))

for i in range(Niter):
    
    kappa_map_gaussian = hp.synfast(interpolate_cl(ell_linear), nside, lmax=3*nside-1, mmax=None, pol=False, pixwin=True)
    
    for j in range(17):
        gpda = hp.gnomview(kappa_map_gaussian, rot=[20*j+75.0,-52.5],reso=4.5, xsize=256, ysize=256,
                                 flip ='geo', return_projected_map =True)
        plt.close()
        gpdb = hp.gnomview(kappa_map_gaussian, rot=[20*j+75.0, 52.5],reso=4.5, xsize=256, ysize=256,
                                 flip ='geo', return_projected_map =True)
        plt.close()
        gpdc = hp.gnomview(kappa_map_gaussian, rot=[20*j+75.0, 0.],reso=4.5, xsize=256, ysize=256,
                                 flip ='geo', return_projected_map =True)
        plt.close()
        
        fft2d = np.fft.fft2(gpda)
        psd1da = radial_profile(np.fft.fftshift(np.real(fft2d*np.conj(fft2d))))
        fft2d = np.fft.fft2(gpdb)
        psd1db = radial_profile(np.fft.fftshift(np.real(fft2d*np.conj(fft2d))))
        fft2d = np.fft.fft2(gpdc)
        psd1dc = radial_profile(np.fft.fftshift(np.real(fft2d*np.conj(fft2d))))
        
        print(i*17+j)
        ps1ds[:,i*17+j] = (psd1da + psd1db + psd1dc) / 3.

    
```

```python
ps1d_previous = fits.getdata('../data/mice/fits_files/mice_ps1d_kappa.fits')
```

```python
ps1d_mice = np.mean(ps1ds,axis=1)

plt.figure()
plt.plot(ps1d_mice, 'r', label='cosmosis')
plt.plot(ps1d_previous, 'b', label='mice')
plt.yscale('log')
plt.legend()
plt.show()
```

```python
# save the averaged power spectrum to a file
np.save('../data/mice/cosmosis_mice_power_spectrum1d.npy', ps1d_mice)
```
