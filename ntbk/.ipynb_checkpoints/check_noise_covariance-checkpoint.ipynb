```python
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt

from astropy.io import fits
from scipy import ndimage
import math

plt.rcParams['figure.figsize'] = (14, 5)
plt.rcParams.update({'font.size': 13})

```

```python
# load survey
# masked
x = fits.open('../xmpl/mice_2389_catalogue_masked_dec8_forth2.fits')

print(x[1].columns.names)

# get shear values from survey
g1 = x[1].data.field('e1_gal_noise').astype(float)
g2 = x[1].data.field('e1_gal_noise').astype(float)


# get galaxies' coordinates
ra=x[1].data.field('ra_gal')
dec = x[1].data.field('dec_gal')

npix = 256

fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
number_map, _, _, cax1 = ax1.hist2d(ra, (np.rad2deg(np.sin(np.deg2rad(dec)))), bins = (npix, npix), cmap = 'inferno')
ax1.set_title('number of background galaxies'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(number_map, interpolation='none', cmap='inferno') #, vmin= 0, vmax = .001 
ax2.set_title('number of background galaxies'), plt.colorbar(cax2, ax=ax2)
plt.show()

fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)

Niter = 200
Nvec = np.zeros((npix**2, Niter)) + 1j * np.zeros((npix**2, Niter))
for m in range(Niter):

    # permutate shear values
    rnx = np.random.permutation(g1.shape[0])
    g1p = g1[rnx]
    g2p = g2[rnx]

    shear_weighted_a, _, _, cax1 = ax1.hist2d(ra, (np.rad2deg(np.sin(np.deg2rad(dec)))), weights = ((g1p)),bins = (npix,npix), cmap = 'inferno')
    shear_weighted_b, _, _, cax2 = ax2.hist2d(ra, (np.rad2deg(np.sin(np.deg2rad(dec)))), weights = ((g2p)),bins = (npix,npix), cmap = 'inferno')

    # this division insert nans because many pixels have zero galaxies
    ni_real = np.divide(shear_weighted_a, number_map, out=np.zeros_like(shear_weighted_a), where=number_map!=0)
    ni_imag = np.divide(shear_weighted_b, number_map, out=np.zeros_like(shear_weighted_b), where=number_map!=0)

    # ni = np.complex(np.reshape(ni_real,-1),np.reshape(ni_imag,-1))

    Nvec[:,m] = np.reshape(ni_real,-1) + 1j * np.reshape(ni_imag,-1)

# compute the covariance matrix 
Sn = np.zeros((npix *npix))#, npix * npix))
for m in range(Niter):
    for i in range(npix * npix):
        Sn[i] += Nvec[i,m].real**2 + Nvec[i,m].imag**2 # np.outer(Nvec[:,m],np.conj(Nvec[:,m])).real # 
Sn = Sn / (Niter - 1)
Sn[Sn==0] = 1e9
```

```python
# np.reshape(Sn,(npix,npix))

# Sn_diag = np.diag(Sn)

Sn_cpp = fits.getdata('../build/Sn.fits')
fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(Sn_cpp, interpolation='none', cmap='inferno', vmin= 0, vmax = .001) # 
ax1.set_title('Sn'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(np.reshape(Sn,(npix,npix)).T, interpolation='none', cmap='inferno', origin='lower', vmin= 0, vmax = .1) # 
ax2.set_title('Sn'), plt.colorbar(cax2, ax=ax2)

plt.show() # 

print(np.max(Sn))
```
