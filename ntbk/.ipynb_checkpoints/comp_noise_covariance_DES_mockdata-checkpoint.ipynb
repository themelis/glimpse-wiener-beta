# Compute the empirical shear noise covariance of the mock DES mice data and wiener filters the DES map

```python

import numpy as np
import random

import os, sys
# import pandas as pd
import healpy as hp
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (16, 6); plt.rcParams.update({'font.size': 13});
```

```python
def make_healpix_map(ra, dec, weights, nside):

    pixels= hp.ang2pix(nside,theta = 0.5*np.pi - np.deg2rad(dec), phi = np.deg2rad(ra))

    bincount = np.bincount(pixels, minlength = hp.nside2npix(nside))
    bincount_weighted = np.bincount(pixels, minlength = hp.nside2npix(nside), weights=weights)

    return np.where(bincount>0.5, bincount_weighted/bincount, hp.UNSEEN)


def shape_noise_realisation(ra, dec, e1_orig, e2_orig, nside):

    gamma1_shuffle = np.copy(e1_orig)
    gamma2_shuffle = np.copy(e2_orig)
    
    random.shuffle(gamma1_shuffle)
    random.shuffle(gamma2_shuffle)
    
    e1_noise = make_healpix_map(ra, dec, gamma1_shuffle, nside)
    e2_noise = make_healpix_map(ra, dec, gamma2_shuffle, nside)

    return e1_noise, e2_noise


```

```python

input_file_des = '../data/mice/fits_files/cat_DES_SV_zmean_final.fits'
hdu_des =  fits.open(input_file_des)
print(hdu_des.info())

des_cat = hdu_des[1].data
print(des_cat.columns)


# code to generate noise realisation
e1_des_noise, e2_des_noise = shape_noise_realisation(des_cat.field('ra_gal'), des_cat.field('dec_gal'),
                                           des_cat.field('e1_gal_sens'), des_cat.field('e2_gal_sens'), 1024)

e1_noise_map = hp.gnomview(e1_des_noise, rot=[+75.0,-52.5],  title='Mask', 
                           min = -.2, max=.2, reso=4.5, xsize=256, ysize=256, flip ='geo', return_projected_map =True)
_ = plt.close()


e2_noise_map = hp.gnomview(e2_des_noise, rot=[+75.0,-52.5],  title='Mask', 
                           min = -.2, max=.2, reso=4.5, xsize=256, ysize=256, flip ='geo', return_projected_map =True)
_ = plt.close()
```

```python
# compute noise covariance from data 

npix = 256
ngal = des_cat.field('ra_gal').shape[0]
Niter = 300 # this will take a while...
Nsvec = np.zeros((npix**2, Niter)) + 1j * np.zeros((npix**2, Niter))
for m in range(Niter):
    
    # code to generate noise realisation
    e1_des_noise, e2_des_noise = shape_noise_realisation(des_cat.field('ra_gal'), des_cat.field('dec_gal'),
                                               des_cat.field('e1_gal_sens'), des_cat.field('e2_gal_sens'), 1024)

    e1_noise_map = hp.gnomview(e1_des_noise, rot=[+75.0,-52.5],  title='Mask', 
                               min = -.2, max=.2, reso=4.5, xsize=256, ysize=256, flip ='geo', return_projected_map =True)
    _ = plt.close()


    e2_noise_map = hp.gnomview(e2_des_noise, rot=[+75.0,-52.5],  title='Mask', 
                               min = -.2, max=.2, reso=4.5, xsize=256, ysize=256, flip ='geo', return_projected_map =True)
    _ = plt.close()
    e1_noise_map = np.where(e1_noise_map.mask==True,.0,e1_noise_map.data)
    e2_noise_map = np.where(e2_noise_map.mask==True,.0,e2_noise_map.data)
    
    # vectorize them
    Nsvec[:,m] = np.reshape(e1_noise_map,-1) + 1j * np.reshape(e2_noise_map,-1)

# compute the covariance matrix in vectorized form
N_diagonal = np.zeros((npix * npix))
for m in range(Niter):
    for i in range(npix * npix):
        N_diagonal[i] += Nsvec[i,m].real**2 + Nsvec[i,m].imag**2 # np.outer(Nvec[:,m],np.conj(Nvec[:,m])).real # 
N_diagonal = N_diagonal / (Niter - 1)

# set high variance to empty pixels
N_diagonal[N_diagonal==0] = 1e9




```

```python
N_theor = np.load('../data/mice/Ncov.npy')

fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(np.reshape(N_diagonal,(npix,npix)), interpolation='none', cmap='inferno',origin='lower',vmax=1) 
ax1.set_title('noise covariance'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.plot(N_diagonal.flatten()[np.where(N_diagonal.flatten() <1)])
plt.show()

fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(np.reshape(N_diagonal,(npix,npix)), interpolation='none', cmap='inferno',origin='lower',vmax=1) 
ax1.set_title('noise covariance'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.scatter(N_diagonal.flatten()[np.where(N_diagonal.flatten() <1)],N_theor.flatten()[np.where(N_theor.flatten() <1)])
plt.show()


```

## wiener filter the shear

```python
# generate the shear values

e1_des = make_healpix_map(des_cat.field('ra_gal'), des_cat.field('dec_gal'), des_cat.field('e1_gal_sens'), 1024)
e2_des = make_healpix_map(des_cat.field('ra_gal'), des_cat.field('dec_gal'), des_cat.field('e2_gal_sens'), 1024)


e1_prj_map = hp.gnomview(e1_des, rot=[+75.0,-52.5],  title='Mask', 
                           min = -.2, max=.2, reso=4.5, xsize=256, ysize=256, flip ='geo', return_projected_map =True)
_ = plt.close()


e2_prj_map = hp.gnomview(e2_des, rot=[+75.0,-52.5],  title='Mask', 
                           min = -.2, max=.2, reso=4.5, xsize=256, ysize=256, flip ='geo', return_projected_map =True)
_ = plt.close()

e1_map = np.where(e1_prj_map.mask==True,.0,e1_prj_map.data)
e2_map = np.where(e2_prj_map.mask==True,.0,e2_prj_map.data)

fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(e1_map, interpolation='none', cmap='inferno',origin='lower') 
ax1.set_title('e1 map'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(e2_map, interpolation='none', cmap='inferno',origin='lower') 
ax2.set_title('e2 map'), plt.colorbar(cax2, ax=ax2)

# plt.show()

e1_prj.shape
```

```python
# use the theoretical noise covariance and power spectrum

Nthr = np.load('../data/mice/Ncov.npy')
ps1d = np.load('../data/mice/projected_mice_power_noisy2.npy')


ps1d.shape
```

```python
# wiener filtering inversion

# define the kappa to shear operator
def H_operator(ka_map, kb_map):
    
    # ka_map and kb_map should be of the same size
    [nx,ny] = ka_map.shape
    
    ka_map_fft = np.fft.fft2(ka_map)
    kb_map_fft = np.fft.fft2(kb_map)
    
    f1, f2 = np.meshgrid(np.fft.fftfreq(ny),np.fft.fftfreq(nx))
    
    p1 = f1 * f1 - f2 * f2
    p2 = 2 * f1 * f2
    f2 = f1 * f1 + f2 * f2
    f2[0,0] = 1 # avoid division with zero
    kafc =  (p1 * ka_map_fft - p2 * kb_map_fft) / f2
    kbfc =  (p1 * kb_map_fft + p2 * ka_map_fft) / f2
    
    g1_map = np.fft.ifft2(kafc).real
    g2_map = np.fft.ifft2(kbfc).real
    
    return g1_map, g2_map


# define the shear to convergence operator
def H_adjoint(g1_map, g2_map):
    
    [nx,ny] = g1_map.shape
    
    g1_map_ifft = np.fft.ifft2(g1_map)
    g2_map_ifft = np.fft.ifft2(g2_map)
    
    f1, f2 = np.meshgrid(np.fft.fftfreq(ny),np.fft.fftfreq(nx))
    
    p1 = f1 * f1 - f2 * f2
    p2 = 2 * f1 * f2
    f2 = f1 * f1 + f2 * f2
    f2[0,0] = 1
    g1fc =  (p1 * g1_map_ifft + p2 * g2_map_ifft) / f2
    g2fc =  (p1 * g2_map_ifft - p2 * g1_map_ifft) / f2
    
    kappa1 = np.fft.fft2(g1fc).real
    kappa2 = np.fft.fft2(g2fc).real
    
    return kappa1, kappa2

def compute_spectrum_map(Px,size):
    power_map = np.zeros((size, size), dtype = float)
    k_map =  np.zeros((size, size), dtype = float)

    for (i,j), val in np.ndenumerate(power_map):

        k1 = i - size/2.0
        k2 = j - size/2.0
        k_map[i, j] = (np.sqrt(k1*k1 + k2*k2))

        if k_map[i,j]==0:
            #print(i,j)
            power_map[i, j] = 1e-15
        else:
            #print(k_map[i, j])
            power_map[i, j] = Px[int(k_map[i, j])]
    return power_map


def prox_wiener_filtering(gamma1, gamma2, Px_map, Ncv):
        
    # initiallize
    nx = gamma1.shape[0]
    xg = np.zeros((nx,nx)) + 1j * np.zeros((nx,nx))
    
    # find the minimum noise variance
    tau = np.min(Ncv)
    
    # set the step size
    eta = 1.83 * tau
    
    # compute signal coefficient
    Esn = eta / Ncv
    
    # calculate the wiener filter coefficients
    Wfc = Px_map / (Px_map + eta)
    
    n_iter = 1000
    for n in range(n_iter):
        t1,t2 = H_operator(xg.real, xg.imag)                              # H * xg
        t1,t2 = H_adjoint(Esn*(gamma1 - t1), Esn*(gamma2 - t2))           # H^T(eta / Sn * (y- H * xg))
        t = xg + (t1 + 1j*t2)                                             # xg + H^T(eta / Sn * (y- H * xg))
        
        xg = np.fft.ifft2(np.fft.fftshift(Wfc * np.fft.fftshift(np.fft.fft2(t)))) # wiener filtering in fourier space
        
    return xg.real, xg.imag

(kpwa,kpwb) = prox_wiener_filtering(e1_map, -e2_map, compute_spectrum_map(ps1d, 256), Nthr)
    
```

```python
fig1, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
cax1 = ax1.imshow(kpwa, interpolation='none', cmap='inferno',origin='upper') # ,vmin=-.015,vmax=.015 
ax1.set_title('wiener estimate e-mode'), plt.colorbar(cax1, ax=ax1)
cax2 = ax2.imshow(kpwb, interpolation='none', cmap='inferno',origin='lower') 
ax2.set_title('wiener estimate b-mode'), plt.colorbar(cax2, ax=ax2)

plt.plot()
```
